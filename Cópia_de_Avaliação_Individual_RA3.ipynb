# **Resposta**:

Para determinar a área sob a curva da função $( y = x^5(2 - x) $) ao longo dos primeiros dois anos, precisamos calcular a integral definida da função de $( x = 0 $) a $( x = 2 $):

$$ \text{Área} = \int_{0}^{2} x^5(2 - x) \, dx $$

Vamos resolver essa integral:

$$ \int_{0}^{2} x^5(2 - x) \, dx = \int_{0}^{2} (2x^5 - x^6) \, dx $$

Para resolver a integral, usamos as regras de integração. Vamos calcular cada termo separadamente:

Primeiro termo:
$$ \int 2x^5 \, dx $$

Usamos a regra de potência para integrar:
$$ \int x^n \, dx = \frac{x^{n+1}}{n+1} $$

Aplicando a regra:
$$ \int 2x^5 \, dx = \frac{2}{6}x^6 = \frac{1}{3}x^6 $$

Segundo termo:
$$ \int -x^6 \, dx $$

$$ \int x^n \, dx = \frac{x^{n+1}}{n+1} $$

Aplicando a regra:
$$ \int -x^6 \, dx = -\frac{1}{7}x^7 $$

Agora, vamos calcular a integral definida de $( x = 0 $) a $( x = 2 $):

$$ \text{Área} = \left[ \frac{1}{3}x^6 - \frac{1}{7}x^7 \right]_{0}^{2} $$

Substituímos $( x = 2 $) e $( x = 0 $):

$$ \text{Área} = \left( \frac{1}{3}(2^6) - \frac{1}{7}(2^7) \right) - \left( \frac{1}{3}(0^6) - \frac{1}{7}(0^7) \right) $$

$$ \text{Área} = \left( \frac{1}{3}(64) - \frac{1}{7}(128) \right) - \left( 0 - 0 \right) $$

$$ \text{Área} = \left( \frac{64}{3} - \frac{128}{7} \right) - \left( 0 - 0 \right) $$

Por fim, encontramos a área:

$$ \text{Área} = \frac{448 - 384}{21} $$

$$ \text{Área} = \frac{64}{21} $$

Portanto, a área sob a curva da função ao longo dos primeiros dois anos é $( \frac{64}{21} $) unidades de área (a unidade de medida depende das unidades de $( x $) e $( y $), que no caso seriam metros e anos, respectivamente).


# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

# **Resposta**:

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def F(x):
    return x**5 * (2 - x)

x = np.linspace(0, 2, 100)
y = F(x)

x_sym = sp.Symbol('x')
area = sp.integrate(x_sym**5 * (2 - x_sym), (x_sym, 0, 2))

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b', label='$F(x) = x^5(2 - x)$')
plt.fill_between(x, y, color='lightblue', alpha=0.5)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Gráfico de $F(x) = x^5(2 - x)$ e Área Sob a Curva')
plt.grid(True)
plt.legend()

plt.text(0.5, 3, f'Área = {area.evalf():.2f}', fontsize=12, ha='center')

plt.show()
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

# **Resposta**:
A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontraremos a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento.

Primeiro, integramos a aceleração para encontrar a velocidade:
\begin{align*}
v(t) &= \int a(t) \, dt \\
&= \int \cos(\pi t) \, dt \\
&= \frac{1}{\pi} \sin(\pi t) + C
\end{align*}

Usamos a velocidade inicial $v(0) = \frac{1}{2\pi}$ para encontrar a constante $C$:
\begin{align*}
\frac{1}{2\pi} &= \frac{1}{\pi} \sin(0) + C \\
\frac{1}{2\pi} &= 0 + C \\
C &= \frac{1}{2\pi}
\end{align*}

Portanto, a função da velocidade é:
$$ v(t) = \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi} $$

Integramos novamente a velocidade para encontrar a posição:
\begin{align*}
d(t) &= \int v(t) \, dt \\
&= \int \left( \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi} \right) \, dt \\
&= -\frac{1}{\pi} \cos(\pi t) + \frac{1}{2\pi} t + C_2
\end{align*}

Usamos a condição inicial $d(0) = 0$ para determinar $C_2$:
\begin{align*}
0 &= -\frac{1}{\pi} \cos(0) + \frac{1}{2\pi} (0) + C_2 \\
0 &= -\frac{1}{\pi} \cdot 1 + 0 + C_2 \\
C_2 &= \frac{1}{\pi}
\end{align*}

Portanto, a função da posição é:
$$ d(t) = -\frac{1}{\pi} \cos(\pi t) + \frac{1}{2\pi} t + \frac{1}{\pi} $$

Para encontrar a distância total percorrida nos primeiros $1,5$ segundos, calculamos $d(1.5)$:
\begin{align*}
d(1.5) &= -\frac{1}{\pi} \cos(\pi \cdot 1.5) + \frac{1}{2\pi} \cdot 1.5 + \frac{1}{\pi} \\
&= -\frac{1}{\pi} \cos(\frac{3\pi}{2}) + \frac{1}{2\pi} \cdot 1.5 + \frac{1}{\pi} \\
&= -\frac{1}{\pi} \cdot 0 + \frac{1.5}{2\pi} + \frac{1}{\pi} \\
&= 0 + \frac{3}{4\pi} + \frac{4}{4\pi} \\
&= \frac{7}{4\pi}
\end{align*}

Aproximando para duas casas decimais, temos:
$$ d(1.5) \approx \frac{7}{4\pi} \cdot \frac{180}{\pi} \, \text{m} \approx 56,55 \, \text{m} $$

Portanto, a distância total percorrida nos primeiros $1,5$ segundos é de aproximadamente $56,55$ metros.


```python
import numpy as np
import matplotlib.pyplot as plt

def a(t):
    return np.cos(np.pi * t)

def v(t):
    return (1/np.pi) * np.sin(np.pi * t) + (1/(2*np.pi))

t_values = np.linspace(0, 1.5, 100)

a_values = a(t_values)
v_values = v(t_values)

plt.figure(figsize=(10, 6))
plt.plot(t_values, a_values, 'b', label='Aceleração ($a(t)$)')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração ($m/s^2$)')
plt.title('Gráfico da Aceleração')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(t_values, v_values, 'r', label='Velocidade ($v(t)$)')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade ($m/s$)')
plt.title('Gráfico da Velocidade')
plt.grid(True)
plt.legend()
plt.show()
```

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de usos de bibliotecas definidas para esta avaliação.  


# **Resposta**:

```python
import numpy as np

def f(x):
    return -x**2 + 4*x + 1

def g(x):
    return -x**3 + 7*x**2 - 10*x + 3

a = 1
b = 2
n = 1000

x_values = np.linspace(a, b, n)

area_f = np.sum(f(x_values) * (b - a) / n)
area_g = np.sum(g(x_values) * (b - a) / n)

diferenca_area = abs(area_f - area_g)

print("A área limitada por f(x) e g(x) no intervalo [1, 2] é:", diferenca_area)
```

# Questão 5 - Vale até 1 ponto

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


Para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $y = \sqrt{x}$, $y = 3$, e o próprio eixo $y$ em torno do eixo $y$, podemos usar o método dos discos.

O método dos discos consiste em integrar a área de seções transversais do sólido, que neste caso são discos circulares.

A função $y = \sqrt{x}$ intersecta a função $y = 3$ em $x = 9$. Assim, para encontrar os limites de integração para $x$, podemos resolver $\sqrt{x} = 3$ para $x$, o que nos dá $x = 9$. Portanto, os limites de integração para $x$ serão de 0 a 9.

A área de um disco é dada por $A = \pi r^2$, onde $r$ é o raio do disco. No nosso caso, o raio $r$ é a distância entre a função $y = \sqrt{x}$ e o eixo $y$, que é simplesmente $y$. Portanto, o raio é $y$.

Então, o volume $V$ do sólido é dado pela integral:

$$V = \int_{0}^{9} \pi (\text{raio})^2 \, dx$$

$$V = \int_{0}^{9} \pi y^2 \, dx$$

Agora, como a função está em termos de $y$, precisamos expressar $y$ em termos de $x$. Sabemos que $y = \sqrt{x}$, então $x = y^2$. Assim, podemos reescrever a integral em termos de $y$:

$$V = \int_{0}^{3} \pi (y^2)^2 \, dy$$

$$V = \int_{0}^{3} \pi y^4 \, dy$$

Agora, podemos resolver essa integral:

$$V = \pi \int_{0}^{3} y^4 \, dy$$

$$V = \pi \left[ \frac{y^5}{5} \right]_{0}^{3}$$

$$V = \pi \left( \frac{3^5}{5} - \frac{0^5}{5} \right)$$

$$V = \pi \left( \frac{243}{5} \right)$$

$$V = \frac{243}{5} \pi$$

Portanto, o volume do sólido é $\frac{243}{5} \pi$ unidades cúbicas.


```python
import sympy as sp

y = sp.Symbol('y')
func = y**4

integral_result = sp.integrate(func, (y, 0, 3))

print("O volume do sólido é:", integral_result * sp.pi)
```

# Questão 6 - Vale até 3 pontos
Para os problemas a e b, determine a média, usando a integração, para a funções nos intervalos especificados.

b. Considere que a posição de um objeto em movimento ao longo do eixo $ x $ é descrita pela função de deslocamento $ x(t) = t^3 - 6t^2 + 9t + 1 $, onde $ x(t) $ está em metros e $t$ está em segundos. Determine a velocidade média do objeto no intervalo de tempo de $ t = 1$ segundo até $t = 4$ segundos. - Vale 2 pontos

 A)

Dada a função $( f(x) = 8x - 3 + 5e^{2-x} $) no intervalo $( [0, 2] $), a média é dada por:
$$ \text{Média} = \frac{1}{b - a} \int_{a}^{b} f(x) \, dx $$

Substituindo os limites de integração e a função \( f(x) \), temos:
$$ \text{Média} = \frac{1}{2 - 0} \int_{0}^{2} (8x - 3 + 5e^{2-x}) \, dx $$

Calculando a integral, temos:
\begin{align*}
\text{Média} &= \frac{1}{2} \left[ \int_{0}^{2} 8x \, dx - \int_{0}^{2} 3 \, dx + \int_{0}^{2} 5e^{2-x} \, dx \right] \\
&= \frac{1}{2} \left[ \left[ 4x^2 \right]_0^2 - \left[ 3x \right]_0^2 + \left[ -5e^{2-x} \right]_0^2 \right] \\
&= \frac{1}{2} \left[ (16 - 0) - (6 - 0) + (5e^0 - 5e^2) \right] \\
&= \frac{1}{2} \left[ 10 - 5e^2 \right] \\
&= 5 - \frac{5}{2}e^2
\end{align*}

Portanto, a média para o problema A) é $( 5 - \frac{5}{2}e^2 $).

 B)

Dada a função de deslocamento $( x(t) = t^3 - 6t^2 + 9t + 1 $) no intervalo de tempo de $( t = 1 $) segundo até $( t = 4 $) segundos, a velocidade média é dada por:
$$ \text{Velocidade Média} = \frac{\Delta x}{\Delta t} $$

Onde $( \Delta x = x(4) - x(1) $) e $( \Delta t = 4 - 1 $). Calculando os valores de $( x(4) $) e $( x(1) $):
$$ x(4) = 4^3 - 6 \cdot 4^2 + 9 \cdot 4 + 1 $$
$$ x(1) = 1^3 - 6 \cdot 1^2 + 9 \cdot 1 + 1 $$

Substituindo na fórmula da velocidade média, temos:
$$ \text{Velocidade Média} = \frac{(4^3 - 6 \cdot 4^2 + 9 \cdot 4 + 1) - (1^3 - 6 \cdot 1^2 + 9 \cdot 1 + 1)}{4 - 1} $$
$$ = \frac{(64 - 96 + 36 + 1) - (1 - 6 + 9 + 1)}{3} $$
$$ = \frac{5}{3} - \frac{5}{3} $$
$$ = 0 $$

Portanto, a velocidade média para o problema b) é 0.



```python
import sympy as sp

x = sp.Symbol('x')
t = sp.Symbol('t')

f_x = 8*x - 3 + 5*sp.exp(2 - x)
x_t = t**3 - 6*t**2 + 9*t + 1

intervalo_a = (0, 2)
intervalo_b = (1, 4)

media_a = (1 / (intervalo_a[1] - intervalo_a[0])) * sp.integrate(f_x, (x, intervalo_a[0], intervalo_a[1]))

x_1 = x_t.subs(t, intervalo_b[0])
x_2 = x_t.subs(t, intervalo_b[1])
media_b = (x_2 - x_1) / (intervalo_b[1] - intervalo_b[0])

print("Problema a:")
print("Média =", media_a)

print("\nProblema b:")
print("Velocidade Média =", media_b)
```

# **Resposta**:

A)


Para a função $f(x) = 4x^3 - 6x^2 + 2x - 5$, vamos calcular a integral indefinida termo a termo:

1. Integral de $4x^3$:

$$\int 4x^3 \, dx = \frac{4}{4}x^4 + C_1 = x^4 + C_1$$

2. Integral de $-6x^2$:

$$\int -6x^2 \, dx = -\frac{6}{3}x^3 + C_2 = -2x^3 + C_2$$

3. Integral de $2x$:

$$\int 2x \, dx = \frac{2}{2}x^2 + C_3 = x^2 + C_3$$

4. Integral de $-5$:

$$\int -5 \, dx = -5x + C_4$$

Agora, combinando todas as integrais, obtemos a família de antiderivadas:

$$F(x) = x^4 - 2x^3 + x^2 - 5x + C$$

Portanto, a família de antiderivadas para $f(x) = 4x^3 - 6x^2 + 2x - 5$ é $F(x) = x^4 - 2x^3 + x^2 - 5x + C$, onde $C$ é a constante de integração.


B)



Para a função $f(x) = \frac{1}{x} + e^x$, vamos calcular a integral indefinida termo a termo:

1. Integral de $\frac{1}{x}$:

$$\int \frac{1}{x} \, dx = \ln|x| + C_1$$

2. Integral de $e^x$:

$$\int e^x \, dx = e^x + C_2$$

Agora, combinando todas as integrais, obtemos a família de antiderivadas:

$$F(x) = \ln|x| + e^x + C$$

Portanto, a família de antiderivadas para $f(x) = \frac{1}{x} + e^x$ é $F(x) = \ln|x| + e^x + C$, onde $C$ é a constante de integração.
